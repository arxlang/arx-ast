# Literals in ASTx

## Introduction

Literals are fixed values embedded directly in source code, representing specific values. They are the raw data given in variables or constants and are not variables themselves but the data assigned to them. Common types of literals found in many programming languages include:

1. **Integer Literals**: Represent whole numbers without any fractional component, e.g., `42`, `-1`, `0`.
2. **Floating-Point Literals**: Represent real numbers with fractional parts, e.g., `3.14`, `-0.01`.
3. **String Literals**: Enclose sequences of characters, typically in quotes, e.g., `"hello"`, `'world'`.
4. **Boolean Literals**: Represent truth values, usually denoted by `true` and `false`.
5. **Null/None Literals**: Represent the absence of a value or a null reference, e.g., `null` in Java, `None` in Python.
6. **Character Literals**: Represent single characters, e.g., `'a'`, `'Z'`.
7. **Array or List Literals**: Define arrays or lists directly, e.g., `[1, 2, 3]` in Python.
8. **Object Literals**: Used in languages like JavaScript to define objects directly, e.g., `{name: "Alice", age: 25}`.

Currently, **ASTx** just support integer literals (8, 16, 32, and 64 bits) and boolean literals, but the support for new ones should be done soon.

Now, let's check how a math operation over literals would look like:

```python
lit_1_a = astx.LiteralInt32(value=1)
lit_1_b = astx.LiteralInt32(value=1)
lit_2 = astx.LiteralInt32(value=2)

(lit_1_a + lit_1_b + lit_2)
```

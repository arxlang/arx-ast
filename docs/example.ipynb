# ASTx Example

```python
from IPython.display import display
import astx
```

```python
module = astx.Module()

a = astx.Variable(name="a", type_=astx.Int32, value=astx.Int32Literal(1))
b = astx.Variable(name="b", type_=astx.Int32, value=astx.Int32Literal(2))
c = astx.Variable(name="c", type_=astx.Int32, value=astx.Int32Literal(4))

lit_1 = astx.Int32Literal(1)

basic_op = lit_1 + b - a * c / a + (b - a / a)

main_proto = astx.FunctionPrototype(
    name="main", args=[], return_type=astx.Int32
)
main_block = astx.Block()
main_block.append(astx.Return(basic_op))
main_fn = astx.Function(prototype=main_proto, body=main_block)

module.block.append(main_fn)
```
